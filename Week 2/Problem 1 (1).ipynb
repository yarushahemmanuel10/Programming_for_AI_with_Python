```python

names = ["Jim", "Hetty", "Kirsten", "Theo", "Henry", "Paul"]
grouped = {} 
""" 
dictionary grouping names based on the first letter 
"H" à "Henry", "Hetty" 
"J" à "Jim" 
"K" à "Kirsten" 
"P" à "Paul" 
"T" à "Theo" 
""" 
for i in names:
    grouped[i[0]] = i

print(grouped)
```
