```python
columns = [ "Age",
"Gender",
"TB Total Bilirubin",
"DB Direct Bilirubin",
"Alkphos Alkaline Phosphotase",
"Sgpt Alamine Aminotransferase",
"Sgot Aspartate Aminotransferase",
"TP Total Protiens",
"ALB Albumin",
"A/G Ratio",
"Class"]
```

```python
from category_encoders.ordinal import OrdinalEncoder
```

```python
maplist = [{
    'col': 'Gender',
    'mapping': {'Male':0, 'Female':1}
}]

oenc = OrdinalEncoder(mapping = maplist)
df = oenc.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.iloc[:, 0:-1]
```

```python
X
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
df['Class'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 23)
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
tree.score(X_test, y_test)
```

```python
import random
random_states = random.sample(range(1, 100), 10)
print(random_states)
```

```python
acc_dt = []
for rs in random_states:
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=rs)
    
    tree = DecisionTreeClassifier(random_state=rs)
    tree.fit(X_train, y_train)
    
    score = tree.score(X_test, y_test)  # returns accuracy
    acc_dt.append(score)
    print(f"Random state {rs} - Accuracy: {score:.3f}")
```

```python
acc_dt
```

```python
acc_mean_dt = sum(acc_dt)/len(acc_dt)
```

```python
print('The average value for scores is:', acc_mean_dt)
```
