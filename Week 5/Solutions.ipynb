```python
import pandas as pd
```

```python
df = pd.read_csv('./ClusterDs1.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(
    df.iloc[:,0],
    df.iloc[:,1],
    c = 'blue',
    marker = 'o',
    edgecolor = 'blue',
    s= 20
    
)
```

```python
from sklearn.cluster import KMeans # required import
import numpy as np
```

```python
km = KMeans(
    n_clusters = 3,
    init = 'random',
    max_iter = 600,
    random_state = 1
    
)
```

```python
labels = km.fit_predict(df)
```

```python
labels
```

```python
plt.scatter(
    df.iloc[labels==0,0],
    df.iloc[labels==0,1],
    s = 50,
    c = 'lightgreen',   # cluster 1
    marker = 's',
    edgecolor = 'black',
    label = 'cluster 1'
)

plt.scatter(
    df.iloc[labels==1,0],
    df.iloc[labels==1,1],
    s = 50,
    c = 'blue',
    marker = 'o',
    edgecolor = 'black',
    label = 'cluster 2'
)

plt.scatter(
    df.iloc[labels==2,0],
    df.iloc[labels==2,1],
    s = 50,
    c = 'red',
    marker = '^',
    edgecolor = 'black',
    label = 'cluster 3'
)

plt.legend()
plt.grid()
```

```python
km.cluster_centers_
```

```python
plt.scatter(
    df.iloc[labels==0,0],
    df.iloc[labels==0,1],
    s = 50,
    c = 'lightgreen',   # cluster 1
    marker = 's',
    edgecolor = 'black',
    label = 'cluster 1'
)

plt.scatter(
    df.iloc[labels==1,0],
    df.iloc[labels==1,1],
    s = 50,
    c = 'blue',
    marker = 'o',
    edgecolor = 'black',
    label = 'cluster 2'
)

plt.scatter(
    df.iloc[labels==2,0],
    df.iloc[labels==2,1],
    s = 50,
    c = 'yellow',
    marker = '^',
    edgecolor = 'black',
    label = 'cluster 3'
)
plt.scatter(
    km.cluster_centers_[:,0],
    km.cluster_centers_[:,1],
    s = 250,
    c = 'red',
    marker = '*',
    edgecolor = 'black',
    label = 'centroid'
)
plt.legend()
plt.grid()
```

```python
# Question 2
import pandas as pd
columns = ['x','y']
df2 = pd.read_csv('ClusterDs2.csv', names = columns)
df2
plt.scatter(
    df2.iloc[:,0],
    df2.iloc[:,1],
    c = 'blue',
    marker = 'o',
    edgecolor = 'blue',
    s= 20
    
)
km2 = KMeans(
    n_clusters = 3,
    init = 'random',
    max_iter = 600,
    random_state = 1
    
)
predicted_labels = km2.fit_predict(df2)
predicted_labels
plt.scatter(
    df2.iloc[predicted_labels==0,0],
    df2.iloc[predicted_labels==0,1],
    s = 50,
    c = 'lightgreen',   # cluster 1
    marker = 's',
    edgecolor = 'black',
    label = 'cluster 1'
)

plt.scatter(
    df2.iloc[predicted_labels==1,0],
    df2.iloc[predicted_labels==1,1],
    s = 50,
    c = 'blue',
    marker = 'o',
    edgecolor = 'black',
    label = 'cluster 2'
)

plt.scatter(
    df2.iloc[predicted_labels==2,0],
    df2.iloc[predicted_labels==2,1],
    s = 50,
    c = 'red',
    marker = '^',
    edgecolor = 'black',
    label = 'cluster 3'
)

plt.legend()
plt.grid()
plt.scatter(
    km.cluster_centers_[:,0],
    km.cluster_centers_[:,1],
    s = 250,
    c = 'red',
    marker = '*',
    edgecolor = 'black',
    label = 'centroid'
)
```

```python
# Question 3
import numpy as np
points = df2.values
centroids= km2.cluster_centers_
distance = np.sqrt(np.sum((points[:, np.newaxis, :] - centroids[np.newaxis, :, :]) ** 2, axis=2))
df2['nearest_centroid'] = np.argmin(distance, axis=1)
np.array_equal(np.array(df2['nearest_centroid']), predicted_labels)
```
