```python
from keras.models import Sequential   # baskets holding the layers of my ANN
from keras.layers import Dense
from keras.utils import plot_model

import numpy as np
import matplotlib.pyplot as plt
```

```python
# The function that we are going to learn
from math import cos as cosine

# Measure performance
import time
```

```python
import numpy as np
```

```python
x = list(range(0,32))

pi_2 = 3.1415926 / 2 

alpha = pi_2/32

x_radians = list(map(lambda x:alpha*x,x))
y = list(map(lambda x:cosine(x), x_radians))
```

```python
x_radians
```

```python
y
```

```python
plt.plot(x_radians,y)
```

```python
xN = list(map(lambda x: x/pi_2, x_radians))
```

```python
xN
```

```python
plt.plot(x_radians,y, label = 'Unnormalized')
plt.plot(xN, y, c= 'r', label= 'Normalized')
plt.title('Normalized Data and Unnormalized Data')
plt.xlabel('x')
plt.ylabel('cosine(x)')
plt.legend()
plt.show()
```

```python
model = Sequential()
model.add(Dense(5, input_dim=1, activation = 'sigmoid'))
model.add(Dense(3, activation = 'sigmoid'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'mean_squared_error', optimizer = 'adam' )
model.summary()
```

```python
start_time = time.time()

h = model.fit(xN , y, verbose = True , epochs = 3000)
```

```python
dt= time.time() - start_time
dt
```

```python
h.history['loss'][2999]
```

```python
def LossHistory(history):
    plt.plot(h.history['loss'], label = 'Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()


LossHistory(h)
```

```python
plt.plot(xN, model.predict(xN), c = 'b', label = 'predictions')
plt.plot(xN, y, c='r', label = 'Actual Values')
plt.xlabel ("xN")
plt.ylabel ("cos(xN)")
plt.legend()
plt.show()
```

```python
plt.plot(xN, model.predict(xN), c = 'b', label = 'predictions')
plt.plot(x_radians, y, c='r', label = 'Actual Values')
plt.xlabel ("x")
plt.ylabel ("cos(x)")
plt.legend()
plt.show()
```
