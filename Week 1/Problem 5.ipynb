```python
game_state = [["x", "o", "x"],
              ["x", "x", "o"],
              ["o", "x", "x"]]
result = "none"

# check for rows
for i in range(3):
    if game_state[i][0] == game_state[i][1] == game_state[i][2]:
        result = game_state[i][0]
        break

# check for columns
for j in range(3):
    if game_state[0][j] == game_state[1][j] == game_state[2][j]:
        result = game_state[0][j]
        break

# check for diagonals
if game_state[0][0] == game_state[1][1] == game_state[2][2]:
    result = game_state[0][0]
elif game_state[0][2] == game_state[1][1] == game_state[2][0]:
    result = game_state[0][2]

print(result)
```
