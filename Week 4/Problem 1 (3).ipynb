```python
import time as t
```

```python
def playMode(series):
    timings = pd.Series()
    for i in range(100):
        start = t.time()
        mode = series.mode()
        end = t.time()
        duration = end - start
        timings.at[i] = duration
    avg_time = timings.mean()    
    print('Average time taken to run .mode(): ',avg_time )
    return mode
```

```python
playMode(df['Age'])
```
