3. Using a lambda, use filter() to accept a list of pairs and accept the pairs 
whose elements are equal. 
 
[[1,1], [2, 5], [4, 4], [9, 9], [5, 3]] Ã  [[1, 1], [4, 4], [9, 9]]

```python
list_1 = [[1,1], [2, 5], [4, 4], [9, 9], [5, 3]] 
```

```python
equal_pairs = filter(lambda x: x[0] == x[1],list_1)
```

```python
print(list(equal_pairs))
```
