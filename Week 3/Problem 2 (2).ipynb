2. Using a lambda, use filter() to accept a list and filter the integers that are 
divisible by 3. Use the modulus operator. 
 
[3, 4, 7, 8, 9, 12, 13] Ã  [3, 9, 12]

```python
list_1 = [3, 4, 7, 8, 9, 12, 13] 
```

```python
filtered = filter(lambda x : x%3 == 0, list_1)
```

```python
print(list(filtered))
```
